```python
import pandas as pd

TrainDF = pd.read_csv('train.csv')
TrainDF.head()
```

```python
X = TrainDF.iloc[:,:-1]
X.head()
y = TrainDF.iloc[:,-1]
y.head()
```

```python
X.shape
```

```python
from sklearn import decomposition

for i in range(1,55):
    pca = decomposition.PCA(n_components= i)
    pca.fit(X)
    print(i," : \t",sum(pca.explained_variance_ratio_))
```

```python
pca = decomposition.PCA(n_components= 9)
pca.fit(X)
X_pca = pca.fit_transform(X)
```

```python
print(X_pca.shape)
```

```python
from sklearn import cross_validation
X_trainPca , X_testPca , y_trainPca , y_testPca = cross_validation.train_test_split(X_pca,y,train_size= 0.9)
print(X_trainPca.shape ,X_testPca.shape)
```

```python
from sklearn import neighbors
clfPca = neighbors.KNeighborsClassifier(n_neighbors=1)
clfPca.fit(X_trainPca,y_trainPca)
accuracyPca = clfPca.score(X_testPca,y_testPca)
print(accuracyPca)
```

```python
"KERAS"

import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pca = decomposition.PCA(n_components= 20)
pca.fit(X)
X_pca = pca.fit_transform(X)


from keras.utils import np_utils

# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, dummy_y, 
                                                    train_size=0.7)


# baseline model
def create_baseline():
    # create model
    model = Sequential()
    model.add(Dense(20, input_dim=20, kernel_initializer='normal', activation='relu'))
    model.add(Dense(10, kernel_initializer='normal', activation='relu'))
    model.add(Dense(7, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = create_baseline()
model.fit(x = X_train ,y = y_train, epochs = 40 , batch_size = 128)
```

```python
preds = model.evaluate(x = X_test , y = y_test)

print()
print ("Loss = " + str(preds[0]))
print ("Test Accuracy = " + str(preds[1]))
```

```python
def create_baselineNoPca():
    # create model
    model = Sequential()
    model.add(Dense(55, input_dim=55, kernel_initializer='normal', activation='relu'))
    model.add(Dense(40, kernel_initializer='normal', activation='relu'))
    model.add(Dense(25, kernel_initializer='normal', activation='relu'))
    model.add(Dense(15, kernel_initializer='normal', activation='relu'))
    model.add(Dense(7, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
preds = model.evaluate(x = X_test , y = y_test)
print()
print ("Loss = " + str(preds[0]))
print ("Test Accuracy = " + str(preds[1]))
```
